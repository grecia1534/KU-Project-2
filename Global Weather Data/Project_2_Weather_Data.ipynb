```python
# Run the full API call and populate the weather  https://weatherstack.com/documentation#php_cur
# Minimum data points to be pulled: country, cloudiness, precipitation, temp.

country_count = 0
current_count = 0

weather_data_json = []
country_list_called = []
cloudiness_list = []
precipitation_list = []
temp_list = []
lat_list = []
lon_list = []
local_time_list = []
feels_like_list = []
wind_speed_list = []
visibility_list = []
    
for country in country_list:

        country_count += 1

        try:
            print(f'Processing record number {country_count} | {country}')


            params = {
                'access_key': weather_api_key,
                'query': country,
                'units': 'f',
            }

            api_result = requests.get('http://api.weatherstack.com/current', params)

            api_response = api_result.json()
            
            weather_data_json.append(api_response)
            
            country_list_called.append(api_response["location"]["country"])
            lat_list.append(api_response["location"]["lat"])
            lon_list.append(api_response["location"]["lon"])
            temp_list.append(api_response["current"]["temperature"])
            feels_like_list.append(api_response["current"]["feelslike"])
            precipitation_list.append(api_response["current"]["precip"])
            cloudiness_list.append(api_response["current"]["cloudcover"])
            visibility_list.append(api_response["current"]["visibility"])
            wind_speed_list.append(api_response["current"]["wind_speed"])
            local_time_list.append(api_response["location"]["localtime"])

            weather_dict = {'Country': country_list_called,
                                'Lat': lat_list,
                                'Lon': lon_list,
                                'Temperature': temp_list,
                                '"Feels Like"': feels_like_list,
                                'Precipitaion': precipitation_list,
                                'Cloudiness': cloudiness_list,
                                'Visiblity': visibility_list,
                                'Wind Speed': wind_speed_list,
                                'Local Time': local_time_list}

        except:
            print('Country not found. Skipping...')
```

```python
# Create a dataframe from the dictionary.

weather_df = pd.DataFrame(weather_dict)
```
